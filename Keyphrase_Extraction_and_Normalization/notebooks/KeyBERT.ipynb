```python
from keybert import KeyBERT
from tqdm import tqdm
```

```python
keywords_list=[]
Ranking_keywords_list=[]
for i in tqdm(text_data):
    kw_model = KeyBERT()
    keywords_high_div = kw_model.extract_keywords(i,keyphrase_ngram_range=(1,2),diversity=0.3,top_n=10)
    keywords_low_div = kw_model.extract_keywords(i,keyphrase_ngram_range=(1,2),diversity=0.01,top_n=10)
    
    print('keywords_high_div',keywords_high_div)
    print('keywords_low_div',keywords_low_div)
    keywords_list.append([i[0] for i in keywords_high_div])
    Ranking_keywords_list.append(keywords_list)
pat_out_data['KeyBERT_Bigram_results']=pd.Series(keywords_list)
# pat_out_data.to_csv('V:/ML_projects/Merckgroup/Projects/Keyword_tagging/output/recent5k_sample_ouptut.csv')
```
